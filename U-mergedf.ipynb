```python
import os
import pandas as pd
```

```python
datas = [x for x in os.listdir() if "results_" in x and ".xlsx" in x]
datas
```

```python
dfs = []
for data in datas:
    dfs.append(pd.read_excel(data, index_col=0, header=[0,1]))

all_data = pd.concat(dfs, ignore_index=True)
```

```python
all_data[all_data[("Graph", "name")] == "graphs/0004_250.graphml"]
```

```python
df_cleaned = all_data.dropna(thresh=len(all_data.columns) - 10)
df = df_cleaned.sort_values(by=("Graph", "name"), ascending=True)
df.reset_index(drop=True, inplace=True)
```
